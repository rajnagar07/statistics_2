```python
# Question 9: Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.
import numpy as np
import matplotlib.pyplot as plt

mu = 50
sigma = 5
n = 1000
data = np.random.normal(mu,sigma,n)  # mu center or mean  sigma = apx itna chota and itna bada n = how many no's
# print(data)
data_mean = np.mean(data)
data_std = np.std(data)
print(data_mean)
print(data_std)

plt.hist(data, bins=30, color="skyblue", edgecolor = 'black' ,density=True)
plt.title("Histogram of Normally Distributed Data (μ=50, σ=5)")
plt.xlabel("Value")
plt.ylabel("Frequency Density")
plt.show()
```

```python
# Question 10: You are working as a data analyst for a retail company. The company has
# collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
#  235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
# ● Explain how you would apply the Central Limit Theorem to estimate the average sales
# with a 95% confidence interval.
# ● Write the Python code to compute the mean sales and its confidence interval.

# The Central Limit Theorem (CLT) says that the sampling distribution of the sample mean
# approaches a normal distribution as the number of samples increases, even if the population distribution is not normal.

import numpy as np
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275,
               240, 255, 235, 260, 245, 250, 225, 270,
               265, 255, 250, 260]
n = len(daily_sales)
sale_mean = np.mean(daily_sales)
sale_std = np.std(daily_sales)

standard_error = sale_std/np.sqrt(n)

# Step 3: t-critical value for df = n-1 = 19 at 95% CI
t_crit = 2.093  # from t-table

# Step 4: Confidence interval
ci_lower = sale_mean - t_crit * standard_error
ci_upper = sale_mean + t_crit * standard_error
print(f"Mean Daily Sales: {sale_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
